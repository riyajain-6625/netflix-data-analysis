```python
n=len(netflix['release_year'])
movies_per_year = netflix['release_year'].value_counts().sort_index().head(20)
movies_per_year.plot(kind='bar', color='skyblue',width=0.5)
plt.title("Number of Movies Released Each Year on Netflix")
plt.xlabel("Year")
plt.ylabel("Number of Movies")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
