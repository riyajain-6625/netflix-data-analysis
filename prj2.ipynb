```python
netflix['date_added'] = pd.to_datetime(netflix['date_added'], errors='coerce')
netflix['year_added'] = netflix['date_added'].dt.year
netflix
```

```python
netflix['cast_count'] = netflix['cast'].str.split(',').str.len()
netflix
```

```python
netflix['category'] = netflix['listed_in'].apply(lambda x: len(str(x).split(',')))
netflix
```

```python
movie_ratings = netflix[netflix['type'] == 'Movie']['rating'].value_counts().head(10)
movie_ratings.plot(kind="bar")
plt.show()
```

```python
tvshow_ratings = netflix[netflix['type'] == 'TV Show']['rating'].value_counts().head(10)
tvshow_ratings.plot(kind="bar")
plt.show()
```

```python
n=len(netflix['release_year'])
movies_per_year = netflix['release_year'].value_counts().sort_index().head(20)
movies_per_year.plot(kind='bar', color='skyblue',width=0.5)
plt.title("Number of Movies Released Each Year on Netflix")
plt.xlabel("Year")
plt.ylabel("Number of Movies")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
